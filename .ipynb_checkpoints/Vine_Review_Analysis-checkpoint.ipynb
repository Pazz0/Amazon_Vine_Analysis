```python
# Analysis on Vine Table CSV
# imports
import pandas as pd
```

```python
df = pd.read_csv("csv_tables/vine_table.csv")
df.head()
```

```python
# filter total_votes
df_votes = df[df["total_votes"] >= 20]
df_votes.head()
```

```python
df_helpful = df_votes[(df_votes["helpful_votes"]/df_votes["total_votes"]) >= .5]
df_helpful
```

```python
# filter for yes vine reviews
df_vine_yes = df_helpful[df_helpful["vine"] == 'Y']
df_vine_yes.head()
```

```python
# filter for no vine reviews
df_vine_no = df_helpful[df_helpful["vine"] == 'N']
df_vine_no.head()
```

```python
# total number, number of 5 star rev, % of 5 star rev (ea type)
# total votes yes vine
sum_yes = df_vine_yes["total_votes"].sum()
sum_yes
# 2754
```

```python
# star rating over 5 yes vine
five_star_yes = df_vine_yes[df_vine_yes["star_rating"] >= 5.0]
five_star_yes["total_votes"].sum()
# 1741
```

```python
# percent of 5 star yes vine
percent_five_yes = five_star_yes["total_votes"].sum() / df_vine_yes["total_votes"].sum() * 100 
percent_five_yes
# 63 %
```

```python
# No vine Dataframes
#total votes no vine
sum_no = df_vine_no["total_votes"].sum()
sum_no
# 749845 
```

```python
# star rating over 5 no vine
five_star_no = df_vine_no[df_vine_no["star_rating"] >= 5.0]
five_star_no["total_votes"].sum()
# 442750
```

```python
# percent of 5 star no vine
percent_five_no = five_star_no["total_votes"].sum() / df_vine_no["total_votes"].sum() * 100 
percent_five_no
# 59 %
```
